### Data Preprocessing 
'Bare Nuclei' column has 16 missing value attributes (?) that are replaced by column mean values. Caegorical variable is not treated or found to make a transformation using OneHotEncoder.

```python
# importing libraries
import numpy as np
import pandas as pd

# importing and inserting header row to the dataset
# replacing '?' values to NaN
missing_values = ["?"]

input_file = "C:/Users/PC/Downloads/breast-cancer-wisconsin.csv"

df = pd.read_csv(input_file, names = ["Sample_code_number","Clump_Thickness","Cell_Size_Uniformity",
                                      "Cell_Shape_Uniformity","Marginal_Adhesion","Single_Epithelial_Cell_Size",
                                      "Bare_Nuclei","Bland_Chromatin","Normal_Nucleoli","Mitoses","Class"], 
                                       na_values = missing_values)

# exploring datset
df
df.info()

# filling missing values with mean value
mean = df['Bare_Nuclei'].mean(skipna=True)
df['Bare_Nuclei'] = df['Bare_Nuclei'].fillna(mean) 
df['Bare_Nuclei'] = df['Bare_Nuclei'].astype(np.int64)
df['Bare_Nuclei']
print(df.isnull().sum())

# X = independent columns, y = dependent column (class)
X = df.iloc[:, :-1].values
y = df.iloc[:, 10].values
```

### Decision Tree on test and training sets
There is no action taken on outliers because they are huge in number. If deleted, vital information is lost.

```python
# splitting dataset into test and training sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 0)

# fitting classifier ito training set (Decision Tree)
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

#### Training Dataset Accuracy - 100%

```python
# training set accuracy for decision tree
classifier.score(X_train, y_train)
```

#### Test Dataset Accuracy - 90.28%

```python
# test set accuracy for decision tree
classifier.score(X_test, y_test)
```

There is a huge difference in the Accuracies of Test and Training datasets because there is a clear implication 
Overfitting in Training data model. To avoid the model to overfit, we need to introduce pruning (can add max_depth 
argument in DecisionTreeClassifier). Overfitting isoccured when the tree is designed so as to perfectly fit all samples in the training data set and continuous tree formation without stopping.

### Kfold Cross Validation with n_splits = 10 

```python
# kfold cross validation
from sklearn.model_selection import KFold, cross_val_score
kf = KFold(n_splits = 10)
kf
```

### Decision Tree with kfold cross validation 
#### Accuracy - 93.13%

```python
# decision tree kfold cross validation accuracy
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
mean_score = cross_val_score(classifier, X, y, cv = kf).mean()
print(mean_score)
```

### Random Forest with kfold cross validation 
#### Accuracy - 95.71%

```python
# random forest kfold cross validation accuracy
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
mean_score = cross_val_score(classifier, X, y, cv = kf).mean()
print(mean_score)
```

### Predictive power
Feature 2 (Cell_Shape_Uniformity) - Highest predictive power with 21.33%, 
Feature 9 (Mitoses) - Lowest predictive power with 1.56%

```python
# to calculate feature importance
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 0)

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
model = classifier.fit(X_train, y_train)
importances = model.feature_importances_
std = np.std([tree.feature_importances_ for tree in model.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

### Rescaling (normalizing) for Naive Bayes and KNN 
Both the models are based on Euclidean distance, so rescaling required to avoid huge differences in data ranges

```python
# rescaling (normalizing) for naive bayes and knn
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled
```

### Naive Bayes with kfold cross validation
#### Accuracy - 92.56%

```python
# naive bayes kfold cross validation accuracy
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
mean_score = cross_val_score(classifier, X_scaled, y, cv = kf).mean()
print(mean_score)
```

### KNN with kfold cross validation
#### Accuracy - 96.56% for K = 5

```python
# knn kfold cross validation accuracy for k= 1-50 values
from sklearn.neighbors import KNeighborsClassifier
accuracy = []
for i in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors=i)
    mean_score = cross_val_score(classifier, X_scaled, y, cv = kf).mean()
    accuracy.append(i)
    accuracy.append(mean_score)
accuracy
```
